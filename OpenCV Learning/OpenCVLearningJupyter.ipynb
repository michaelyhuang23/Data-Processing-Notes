```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Reading Image

```python
img = cv2.imread('yoda.jpg',0)
# the second argument "flag" indicate the type of image read in. There are cv2.IMREAD_GRAYSCALE (correspond to 0), cv2.IMREAD_COLOR (correspond to 1) and cv2.IMREAD_UNCHANGED (correspond to -1). UNCHANGED reads in the alpha channel in addition to the color channels. The Alpha channel indicates opaqueness. 
```

```python

cv2.imshow('YODA',img)
# first argument is name
cv2.waitKey(0)
# waits for any key to be pressed in the case of 0 (or you could specify key by their integer value)
cv2.destroyAllWindows()
# kills the windows showing images/videos/etc
# this doesn't work well because of notebook environment
```

```python
# or use matplotlib to show image
plt.imshow(img,cmap='gray', interpolation='bicubic')
# cmap = colorMap. It's basically the filter
# interpolation is a method for predicting the pixel value at a specific pixel based on its surrounding pixels. This is crucial obviously when we resize the image since there might not be originally a pixel on the new pixel position. Bicubic is an interpolation algorithm that predicts the value of a pixel by looking at the cubic curves formed by its adjacent pixels. bilinear of course is only a linear approximation. Bi- of course means two dimensional.
plt.show()
```

Writing Image

```python
cv2.imwrite('yoda_copy.jpg',img)
```

Image/Video Capture with webcam

```python
cap = cv2.VideoCapture(0)
# the input integer indicate the number id of the camera/input source it should access. Most computer has 0 (the first) set as webcam
ret, frame = cap.read()
# ret returns whether or not there's a next frame/feed. frame is basically the frame of picture read-in. 
while ret:
    ret, frame = cap.read()
    cv2.imshow('frame', frame)

cap.release()
cv2.destroyAllWindows()
# currently the camera is being used by python. To allowed the camera to be used elsewhere and to basically stop waiting/checking/reading-in an input, we need to release it.
```
